# Clean the downloaded Data
The species names from GBIF where not super clean

```python
import pandas as pd
```

```python
data = pd.read_csv("MammalsOfGermany.csv")
data.head()
```

```python
data["scientificGenus"] = data["scientificName"].str.split().str[0]
data["scientificSpecies"] = data["scientificName"].str.split().str[1]
data["scientificSpecies"] = data["scientificSpecies"].str.strip(',')
data["scientificName"] = data["scientificGenus"] + " " + data["scientificSpecies"]
data.head()
```

```python
data = data.drop(["scientificGenus", "scientificSpecies"], axis=1)
```

```python

data = (
    data.groupby(['order', 'family', 'scientificName'], as_index=False)
        .agg({'vernacularName': lambda x: ', '.join(sorted(set(x.dropna())))})
)
data.head()
len(data)
```
